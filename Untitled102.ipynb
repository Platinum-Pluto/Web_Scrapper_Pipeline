```python
import asyncio
import pandas as pd
from crawl4ai import AsyncWebCrawler
from bs4 import BeautifulSoup
import nest_asyncio

# Apply nest_asyncio if running in an environment with an active loop (e.g., Jupyter)
nest_asyncio.apply()
links = []
async def scrape_expomed():
    base_url = "https://expomedistanbul.com/katilimci-listesi?participantFullList=true"

    async with AsyncWebCrawler(headless=True) as crawler:
        # Load the main list page
        main_page = await crawler.arun(url=base_url)

        # Use BeautifulSoup to parse the HTML content
        main_soup = BeautifulSoup(main_page.html, "html.parser")
        # Use the provided CSS selector to select all "GÖRÜNTÜLE" buttons
        company_anchor_selector = "#katilimcilar-listesi > div > table > tbody > tr td.hidden-print > a"
        company_links_elements = main_soup.select(company_anchor_selector)

        data_list = []

        for idx, anchor in enumerate(company_links_elements):
            relative_url = anchor.get("href", "")
            if not relative_url:
                continue

            # Construct the full URL for the company details page
            full_url = relative_url if relative_url.startswith("http") else f"https://expomedistanbul.com/{relative_url}"
            links.append(full_url)
            print(f"Scraping company {idx+1}: {full_url}")


# In environments with an active event loop (e.g., Jupyter), await the function directly:
await scrape_expomed()
```

```python
# prompt: store all the values from list links to a excel file

import pandas as pd

# Assuming 'links' list is populated from the previous code
# Create a DataFrame from the links
df = pd.DataFrame({'Links': links})

# Save the DataFrame to an Excel file
df.to_excel('links.xlsx', index=False)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import time
from tqdm import tqdm  # Import tqdm for progress bar

# Read the input Excel file
df = pd.read_excel("/content/links.xlsx")  # Ensure this file is in your working directory

# Prepare lists to store extracted data
company_names = []
addresses = []
phone_numbers = []
websites = []

# Initialize tqdm for showing progress in the loop
for url in tqdm(df['Links'], desc="Processing URLs", unit="url"):
    try:
        # Get the HTML content
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # Raise an error for bad responses
        html = response.content

        # Parse the HTML with BeautifulSoup
        soup = BeautifulSoup(html, "html.parser")

        # Extract company name and address from <div class="table-block-content">
        divs = soup.find_all("div", class_="table-block-content")
        if len(divs) >= 2:
            company_text = divs[0].get_text(strip=True)
            address_text = divs[1].get_text(strip=True)
        else:
            company_text = None
            address_text = None

        # Extract phone number: find <a> tag with href starting with "tel:"
        phone_tag = soup.find("a", href=re.compile(r"^tel:"))
        phone_text = phone_tag.get_text(strip=True) if phone_tag else None

        # Extract website URL: find <a> tag with href that starts with "http" (excluding tel links)
        website_tag = soup.find("a", href=re.compile(r"^(https?)://"))
        website_text = website_tag['href'] if website_tag else None

        # Append the results to the lists
        company_names.append(company_text)
        addresses.append(address_text)
        phone_numbers.append(phone_text)
        websites.append(website_text)

        # Optional: Pause briefly to be polite to the server
        time.sleep(1)

    except Exception as e:
        print(f"Error processing {url}: {e}")
        company_names.append(None)
        addresses.append(None)
        phone_numbers.append(None)
        websites.append(None)

# Create a DataFrame with the extracted data
output_df = pd.DataFrame({
    "Link": df['Links'],
    "Company Name": company_names,
    "Address": addresses,
    "Phone": phone_numbers,
    "Website": websites
})

# Write the results to a new Excel file
output_df.to_excel("/content/output.xlsx", index=False)

print("Data extraction complete. Output saved to output.xlsx")
```
